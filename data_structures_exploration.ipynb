# Get comfortable with some basic data structures

### Lists

```python
arr = ['Dog','Cat','Mouse','Bird','Fish']
len(arr)
```

```python
arr.append('Turtle')
print(arr)
```

```python
arr.pop() # removes an element, default is the last element
print(arr)
```

```python
arr.sort() # sorts the list in place, meaning it modifies the original list
arr
```

```python
sorted_arr = sorted(arr) # creates a new sorted list, leaving the original list unchanged
print(sorted_arr)
print(arr)
```

### Dicts

```python
d = {'Arizona': 'Phoenix', 'California': 'Sacramento', 'Colorado': 'Denver', 'Florida': 'Tallahassee', 'Nevada': 'Carson City'}
d['Arizona'] # access value by key
```

```python
d['Texas'] = 'Austin' # add a new key-value pair to the dictionary
print(d)
```

```python
d.pop('Florida') # removes the key-value pair with the specified key and returns the value
print(d)
```

```python
d.get('Florida', 'Not Found') # returns the value for the specified key if it exists, otherwise returns the default value provided as the second argument
```

```python
print(d.keys()) # returns a view object that displays a list of all the keys in the dictionary
print(d.values()) # returns a view object that displays a list of all the values in the dictionary
print(d.items()) # returns a view object that displays a list of all the key-value pairs in the dictionary as tuples
```

### Sets

```python
seen = set()
seen.add('apple')
seen.add('banana')
seen.add('banana') # adding a duplicate element to a set does not change the set, as sets only store unique elements
print(seen)
```

```python
if 'apple' in seen:
    print('Found apple in the set!')
else:
    print('Apple not found in the set.')
```

### Counters and defaultdicts from the collections module

```python
from collections import Counter, defaultdict
words = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
word_count = Counter(words) # creates a Counter object that counts the occurrences of each element in the list
print(word_count)
```

```python
for word, count in word_count.items():
    print(f'{word}: {count}')
```

```python
graph = defaultdict(list)
graph['A'].append('B')
graph['A'].append('C')
graph['B'].append('A')
graph['B'].append('D')
print(graph)
```

```python
items = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']

counts = {}

for x in items:
    if x not in counts:
        counts[x] = 0
    counts[x] += 1

print(counts)
```

```python
counts = defaultdict(int)

for x in items:
    counts[x] += 1

print(counts)
```

### Queue for BFS

```python
from collections import deque
start = 'A'
neighbor = 'B'
neighbor2 = 'C'
q = deque([start])
print('Queue:',q)
q.append(neighbor)
print('Queue:',q)
node = q.popleft()
print('Node:',node)
q.append(neighbor2)
print('Queue:',q)
```

### Graphs

```python
graph = {
    "A": ["B", "C"],
    "B": ["A", "D", "E"],
    "C": ["A", "F"],
    "D": ["B"],
    "E": ["B", "F"],
    "F": ["C", "E"]
}
graph
```

```python
# Print each node and its neighbors
for node, neighbors in graph.items():
    print(node, "->", neighbors)
```

```python
# convert an array of edges into an adjacency list representation of a graph
edges = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('E', 'F')]
edges = [[0, 1], [0, 2], [1, 3], [1, 4], [2, 5], [4, 5]]
graph = defaultdict(list)
for u, v in edges:
    graph[u].append(v)
    graph[v].append(u) # assuming it's an undirected graph
graph
```

```python
n = 6
M = []
for i in range(n):
    M.append([0] * n) # create a row of zeros and append it to the matrix
M
```

```python
for u,v in edges:
    M[u][v] = 1 # mark the presence of an edge from u to v
M
```
