# Get comfortable with some basic data structures

```python
s = "(])"
```

```python
def isValid(self, s: str) -> bool:
    stack = []
    pairs = {
        ")": "(",
        "]": "[",
        "}": "{",
    }

    for char in s:
        print(char)
        if char in "([{":
            stack.append(char)
        else:
            if not stack or stack[-1] != pairs[char]:
                print(stack)
                print(pairs[char])
                return False
            stack.pop()

    return len(stack) == 0
```
