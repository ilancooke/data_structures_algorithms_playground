# Get comfortable with some basic data structures

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
node3 = ListNode(4)
node2 = ListNode(2, node3)
node1 = ListNode(1, node2)
head = node1
```

```python
current = head
while current:
    print(current.val)
    current = current.next
```
